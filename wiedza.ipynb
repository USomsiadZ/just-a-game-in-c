# Kolorki asci
ESC [ 48 ; 2 ; R ; G ; B m ]
```c
printf("\033[48;2;%d;%d;%dm  \033[0m", r, g, b); // RGB
```
ESC -> tło -> reset tła
```c
printf("\033[41m  \033[0m");   // Czerwony kwadrat (tło)
```

````c
//Przykład użycia
#include <stdio.h>

void pixel(int r, int g, int b) {
    printf("\033[48;2;%d;%d;%dm  \033[0m", r, g, b);
}

int main() {
    pixel(255, 0, 0);   // czerwony
    pixel(0, 255, 0);   // zielony
    pixel(0, 0, 255);   // niebieski
    pixel(255, 255, 0); // żółty
    pixel(255, 0, 255); // magenta
    pixel(0, 255, 255); // cyjan
    pixel(255, 255, 255); // biały
    printf("\n");

    return 0;
}
````
